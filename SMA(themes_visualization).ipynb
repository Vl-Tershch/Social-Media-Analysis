```python
stopwords = set(nltk_stopwords.words('russian'))

tf_vectorized = CountVectorizer( stop_words=stopwords,
                                    max_df = 0.5,
                                    min_df = 1)

dtm_tf = tf_vectorized.fit_transform(json_data)
tfidf_vectorizer = TfidfVectorizer(**tf_vectorized.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(json_data)
# for TF DTM
lda_tf = LatentDirichletAllocation(n_components=2, random_state=0)
lda_tf.fit(dtm_tf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=2, random_state=0)
lda_tfidf.fit(dtm_tfidf)
vis_data = pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorized)
pyLDAvis.display(vis_data)
```
